## Implementación Random Forest

```python
df_copy.groupby('classALeRCE').count()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

scaler = StandardScaler()
Le = LabelEncoder()
X = df_copy.drop(columns = ['classALeRCE', 'oid'])
X = scaler.fit_transform(X)
y = df_copy['classALeRCE'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
from imblearn.over_sampling import SMOTE
sm= SMOTE(sampling_strategy = {'SLSN': 600 , 'SNII': 356*3, 'SNIa': 1394, 'SNIbc': 96*8})
X_train_smote, y_train_smote = sm.fit_resample(X_train, y_train)
X_train, y_train = X_train_smote, y_train_smote
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier as RandomForestClassifier

feature_names = [f"feature {i}" for i in range(X.shape[1])]
forest = RandomForestClassifier(n_estimators=50,
    criterion='gini',
    max_depth=30,
    max_features='sqrt',
    n_jobs=-1,
    class_weight='balanced')
forest.fit(X_train, y_train) 
```

```python
from sklearn import metrics
import seaborn as sns
predict = forest.predict(X_test)
matriz = metrics.confusion_matrix(y_test, predict, normalize='true')

# plt.figure(figsize=(16,9))
sns.heatmap(matriz, annot=True)
plt.title('Matriz de confusión 109 features Random Forest')
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import classification_report
y_pred_test = forest.predict(X_test)
accuracy_score(y_test, y_pred_test)
a = metrics.classification_report(y_test, y_pred_test)
print(a)
```

# **MLP**

```python
#librerias necesarias para Deep Learning segun lo visto en el auxiliar de metodos de clasificacion.
import tensorflow as tf
from tensorflow import keras
from tensorflow.python.keras.utils.vis_utils import plot_model
from keras.callbacks import EarlyStopping
import tensorflow as tf
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Softmax
from keras.utils import to_categorical
from keras.layers import Dropout
```

```python

#Creamos el primer modelo de red neuronal usando la libreria keras, segun lo visto en el auxiliar, con 5 capas ocultas, una capa de entrada y una de salida.
model = Sequential()
model.add(Dense(30, activation= 'relu', input_dim=30)) #Usamos funcion de activacion relu porque es la mas usada en redes neuronales en las capas ocultas
model.add(Dense(40, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(4, activation='sigmoid')) #Ya que buscamos clasificar en 4 categorias, usamos la funcion de activacion sigmoid en la capa de salida
```

```python
#Compilamos nuestro modelo, donde usamos como funcion de perdida la funcion categorical_crossentropy, ya que es la mas usada para clasificacion multiclase, en cojunto con
#el optimizador adam, que es el que mejor desempeño tuvo en este laboratorio.
model.compile(loss= 'categorical_crossentropy', optimizer='sgd', metrics=['accuracy']) 
model.summary()
```

```python
#Compilamos nuestro modelo, donde usamos como funcion de perdida la funcion categorical_crossentropy, ya que es la mas usada para clasificacion multiclase, en cojunto con
#el optimizador adam, que es el que mejor desempeño tuvo en este laboratorio.
#model.compile(loss= 'binary_crossentropy', optimizer='sgd', metrics=['accuracy']) 
#model.summary()
```

```python
print('train set shape',X_train.shape)
print('test set shape',X_test.shape)
print('train label shape',y_train.shape)
print('test label shape',y_test.shape)
```

```python
#Ahora convertiremos nuestras variables objetivo en categorias, para poder entrenar nuestro modelo, puesto que keras no acepta variables numericas como variables objetivo para clasificacion.
y_train = pd.get_dummies(y_train, columns = ['classALeRCE'])
y_test = pd.get_dummies(y_test, columns = ['classALeRCE'])
```

```python
#Ahora veamos como quedo nuestro conjunto y_train
y_train
```

```python
#Ahora entrenaremos nuestro modelo, con 100 epocas, y un batch size de 100, para ver que tan bien se comporta nuestro modelo.
max_epochs = 100
history = model.fit(X_train, y_train, epochs=max_epochs, validation_data=(X_test, y_test), batch_size=100)
```

```python
#Ahora veamos como se comporta nuestro modelo, para lo que graficaremos la funcion de perdida y la funcion de accuracy.
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
#Ahora graficaremos la funcion de perdida y la funcion de accuracy para ver como evolucionan en el tiempo
#Notamos que nuestro modelo alcanza un accuracy de 0.53, lo cual no es muy bueno, pero cumple con ser mayor a 0.5, por lo que es un modelo aceptable, al igual que el modelo anterior.
#En cuanto a la funcion de perdida, notamos que esta disminuye con el tiempo, lo cual es bueno, ya que quiere decir que nuestro modelo esta aprendiendo y no esta cayendo en overfitting.
figura, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12,6))

n_epochs = len(acc)

ax1.plot(range(n_epochs), acc, label='Training accuracy')
ax1.plot(range(n_epochs), val_acc, label='Validation accuracy')
ax1.set_title('Accuracy')
ax1.legend()

ax2.plot(range(n_epochs), loss, label='Training loss')
ax2.plot(range(n_epochs), val_loss, label='Validation loss')
ax2.set_title('Loss')
ax2.legend()
```
